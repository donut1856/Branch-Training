### 替代Vlookup

```python
#可以高效替代excel中的vlookup
#pd.merge可以实现数据表合并

teller = pd.DataFrame({'姓名':['唐三藏','孙悟空','猪八戒','白龙马','沙悟净'],
                       '柜员号':['9880100','9881000','9881900','7072612','6422733']
                      }
)
teller
```

```python
df['最后维护柜员'] = df['最后维护柜员'].astype('str')

vlookup = pd.merge(df, teller
                    ,how='right'
                    ,left_on = '最后维护柜员'
                    ,right_on = '柜员号'
                     )                    #how参数决定以何种方式进行合并，默认为inner
                                          #inner以交集方式合并
                                          #left以左表为主表合并右表数据，最终结果行数必定大于等于左表
                                          #right以右表为主表合并左表数据，最终结果行数必定大于等于右表
                                          #outer以并集方式合并
```

# 数据透视表

<img src="./数据透视表参数.jpg" style="zoom:80%" />

```python
#pd.pivot_table函数可以获得数据透视表

mps = pd.pivot_table(df_new
               ,index= ['归属行名','手机号']
               ,values = ['客户号']
               ,aggfunc = ['count'] #替代方式len，np.count_nonzero（非零计数） 
               ,fill_value = 0
               ,margins = True
               ,margins_name = '总计'    
              ).reset_index ()
mps
```

```python
mps_new = pd.pivot_table(df_new
               ,index= ['归属行名','手机号']
               ,values = ['客户号','月收入']
               ,aggfunc = {'客户号':'count', '月收入':'sum'}
               ,fill_value = 0
               ,margins = True
               ,margins_name = '总计'    
              )
mps_new
```

```python
cs = pd.pivot_table(df_new
               ,index= ['归属行名','创建机构名称','分级结果']
               ,values = ['客户号']
               ,aggfunc = 'count' #替代方式len，np.count_nonzero（非零计数） 
               ,fill_value = 0
               ,margins = True
               ,margins_name = '总计'    
               )
cs
```

```python
#增加百分比

cs['所占本行百分比'] = (cs['客户号']/cs.groupby(['归属行名','创建机构名称'])['客户号'].sum()*100).astype(str)+'%'
cs = cs.reset_index()
cs
```
